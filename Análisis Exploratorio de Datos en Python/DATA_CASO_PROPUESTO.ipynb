## Estudio de precio por cuidad y tipo de inmueble

Siempre y cuando la muestra total sea representativa (mas de 15 muestras)

```python
norepre = dict()
for tipoinmueble in df['Tipo de Inmueble'].unique():
    muestras = df[df['Tipo de Inmueble'] == tipoinmueble].shape[0]
    if muestras < 11:
        norepre[tipoinmueble] = muestras
        continue
    fig = df[df['Tipo de Inmueble'] == tipoinmueble].boxplot("Precio", by="Departamento")
    plt.ylabel('precio')
    plt.suptitle('')
    plt.xticks(rotation=75);
    plt.title(f'Tipo inmueble {tipoinmueble} sobre {muestras} muestras')
print('Inmuebles no representativos por categora')
pd.DataFrame({'Tipo inmueble':norepre.keys(),'muestras':norepre.values()})
```

```python
# se identifica la cantida de diferenes departamentos en la muestra
def showbycity(departamento, minimo = 11):
    dpto = departamento.upper()
    muestras = df[df['Departamento'] == dpto].shape[0]
    if muestras < minimo:
        return (dpto, muestras)
    fig = df[df['Departamento'] == dpto].boxplot("Precio",by="Ciudad")
    plt.ylabel('Precio')
    plt.xticks(rotation=45);
    plt.suptitle('')
    plt.title(f'{dpto} valor por condicin, {muestras} muestras')
```

## Estudio comparativo de variacin de precios por cuidad

```python
# se identifican todas las cuidades
res = dict()
for departamento in df['Departamento'].unique():
    val = showbycity(departamento, minimo=10)
    if val is None:
        continue
    res[val[0]] = val[1]
print('Omitidos en este punto por cantidad de muestras insuficientes')
pd.DataFrame({'Departamento': res.keys() , 'Muestras':res.values()})
```

# Realice histogramas de frecuencia

Se graficarn:

 Cantidad de reportes por estrato

 Cantidad de reportes por departamento

 Rango de valores de vivienda.

```python
# se define el rango de valores a graficar
from math import ceil
bins = 100
minimo = df['Precio'].min(axis=0);
maximo = df['Precio'].max(axis=0);
```

```python
def get_hist_values(data, minimo, maximo, bins):
    values = list();
    delta = ceil((maximo -minimo)/bins);
    initvalue = minimo;
    endvalue = minimo + delta;
    names = []
    inits = []
    ended = []
    cantidad = []
    k=1
    while initvalue < maximo:
        localmax = initvalue + delta + 0.01
        names.append(k)
        inits.append(initvalue)
        ended.append(localmax)
        cantidad.append(df[(df['Precio'] >= initvalue) & (df['Precio'] < localmax)].shape[0])
        initvalue = localmax
        k +=1
    print('Datos para construir el histograma');
    res = pd.DataFrame({'id': names, 'cantidad': cantidad, 'inicial': inits,'final':ended});
    return res
```

```python
res = get_hist_values(df, minimo, maximo, bins)
res.head(20)
```

Con base en el punto anterior se nota que la mayor cantida de datos, 429, est hasta el valor de 76950021, por lo que se ajusta este como el valor lmite y se corre nuevamente la funcin.

```python
res = get_hist_values(df, minimo, 76950021 , bins=200)
res.head(20)
#ax.set_xticklabels(res['inicial'], rotation='vertical', fontsize=10);
# se procede a graficar
```
