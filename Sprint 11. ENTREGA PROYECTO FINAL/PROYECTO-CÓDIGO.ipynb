```python
# comprobación del porcentaje/cantidad de valores atípicos:

for i in seleccion_númericas.columns:
  seleccion_númericas.sort_values(by=i, ascending=True, na_position='last')
  q1, q3 = np.nanpercentile(seleccion_númericas[i], [25,75])
  iqr = q3-q1
  lower_bound = q1-(1.5*iqr)
  upper_bound = q3+(1.5*iqr)
  outlier_data = seleccion_númericas[i][(seleccion_númericas[i] < lower_bound) | (seleccion_númericas[i] > upper_bound)] #creando una serie de datos atípicos
  perc = (outlier_data.count()/seleccion_númericas[i].count())*100
  print('Los valores atípicos en %s son %.2f%% con recuento %.f' %(i, perc, outlier_data.count()))
```

```python
from scipy.stats import chi2_contingency

# Crear una tabla de contingencia para la distribución de la población por sexo
tabla_contingencia = pd.crosstab(df_processed['Sexe'], columns='count')

# Realizar la prueba de chi-cuadrado
chi2, p_value, dof, expected = chi2_contingency(tabla_contingencia)

# Imprimir los resultados
print('Estadístico chi-cuadrado:', chi2)
print('Valor p:', p_value)
```

En esta prueba, la hipótesis nula (H0) es que no hay diferencia significativa en la distribución de la población por sexo. Si el valor p es menor que un umbral de significancia predefinido (por ejemplo, 0.05), entonces se rechaza la hipótesis nula y se concluye que hay diferencias significativas en la distribución de la población por sexo.

# Top 5 Distritos y Barrios con mayor número de Extranjeros

```python
# Filtramos el conjunto de datos para incluir solo las filas con nacionalidad extranjera
df_extranjeros = df_processed[df_processed['Nacionalitat'] != 'Espanyola']

# A continuación, podemos calcular el número total de extranjeros por barrio y distrito utilizando el método groupby y el método sum:
# Agrupamos los datos por distrito y barrio y calculamos la suma de la columna Nombre
extranjeros_por_barrio = df_extranjeros.groupby(['Codi_Districte', 'Nom_Districte', 'Codi_Barri', 'Nom_Barri'])['N.Habitantes'].sum().reset_index()

# Agrupamos los datos por distrito y calculamos la suma de la columna Nombre
extranjeros_por_distrito = extranjeros_por_barrio.groupby(['Codi_Districte', 'Nom_Districte'])['N.Habitantes'].sum().reset_index()

# Seleccionamos los Top 5 distritos con mayor número de extranjeros
top_distritos = extranjeros_por_distrito.nlargest(5, 'N.Habitantes')

# Seleccionamos los Top 5 barrios con mayor número de extranjeros
top_barrios = extranjeros_por_barrio.nlargest(5, 'N.Habitantes')

# Creamos un gráfico de barras horizontal de los Top 5 distritos con mayor número de extranjeros
plt.barh(top_distritos['Nom_Districte'], top_distritos['N.Habitantes'])
plt.xlabel('Número de extranjeros')
plt.ylabel('Distrito')
plt.title('Top 5 distritos con mayor número de extranjeros')
plt.show()

# Creamos un gráfico de barras horizontal de los Top 5 barrios con mayor número de extranjeros
plt.barh(top_barrios['Nom_Barri'], top_barrios['N.Habitantes'])
plt.xlabel('Número de extranjeros')
plt.ylabel('Barrio')
plt.title('Top 5 barrios con mayor número de extranjeros')
plt.show()
```
