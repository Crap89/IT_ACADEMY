```python
df2.info()
```

```python
# Resumen Estadístico General:

df2.describe()
```

```python
# Datos faltantes por columna.

df2.isnull().sum()
```

```python
# Se Reemplaza los datos faltantes con el promedio de cada columna:

reemplazar_datos= df2.fillna(value=df2.mean())
reemplazar_datos
```

```python
# Se verifica que no hayan quedado datos nulos:

reemplazar_datos.isnull().sum()
```

```python
# Crea nuevas columnas (velocidad media del vuelo, si ha llegado tarde o no...)

reemplazar_datos['Velocidad media del vuelo'] = reemplazar_datos['Distance'] / reemplazar_datos['ActualElapsedTime']
reemplazar_datos['Si ha llegado tarde o no'] = reemplazar_datos['LateAircraftDelay']>30.789474
reemplazar_datos
```

```python
# Haz una tabla de las aerolíneas con mayores retrasos acumulados.

mayores_retrasos= reemplazar_datos.groupby('UniqueCarrier')['WeatherDelay', 'NASDelay', 'LateAircraftDelay' ].max().sort_values('LateAircraftDelay'
                                           ,ascending=False).head(10)
mayores_retrasos
```

```python
# Gráfico:

mayores_retrasos.plot(kind='bar')
plt.title('Aerolíneas con Mayores Retrasos Acumulados')
plt.xlabel('Operadores')
plt.ylabel('Retraso (min)')

plt.legend(title='TIPOS DE RETRASOS', bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

```python
# ¿Cuáles son los vuelos más largos?

vuelos_largos= reemplazar_datos.groupby('UniqueCarrier')['Distance'].max().sort_values(ascending=False).head(10)
vuelos_largos
```

```python
# Gráfico:

vuelos_largos.plot(kind='bar')
plt.title('Vuelos más Largos')
plt.xlabel('Operadores')
plt.ylabel('Distancia de Vuelo')

plt.show()
```

```python
# ¿Y los más atrasados?

vuelos_atrasados= reemplazar_datos.groupby('UniqueCarrier')['LateAircraftDelay'].max().sort_values(ascending=False).head(10)
vuelos_atrasados
```

```python
vuelos_atrasados.plot(kind='bar')
plt.title('Vuelos más Atrasados')
plt.xlabel('Operadores')
plt.ylabel('Retraso en Minutos')

plt.show()
```
